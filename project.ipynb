```
import pandas as pd
from random import randint, uniform, choice
from faker import Faker

fake = Faker()

def input_data(x):
    # List of product names for the accessories category
    accessories = ['belt', 'earring', 'glasses', 'glove', 'handbags', 'sun glasses', 'hat', 'watch', 'ring', 'necktie', 'helmet']

    # List of brand names for the accessories category
    brands = ['Gucci','fossil', 'Prada','Kate Spade','Tory Burch','Bvlgari','Cartier','Swarovski','Ray-Ban','Oakley', 'Louis Vuitton', 'Chanel','Michael Kors','Coach', 'Hermès', 'Fendi', 'Balenciaga', 'Dior', 'Versace', 'Burberry']

    # pandas dataframe
    data = pd.DataFrame()
    for i in range(0, x):
        data.loc[i, 'product_name'] = choice(accessories)
        data.loc[i, 'Price'] = round(uniform(100.0, 10000.0), 2)
        data.loc[i, 'Description'] = fake.sentence()
        data.loc[i, 'Category'] = 'accessories'
        data.loc[i, 'StockQuantity'] = randint(1, 100)
        data.loc[i, 'Brand'] = choice(brands)
        data.loc[i, 'Rating'] = round(uniform(1.0, 5.0), 1)
        data.loc[i, 'Warehouse'] = fake.city()
    return data

# Example usage
input_data(10)
```
